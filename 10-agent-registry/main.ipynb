## What is the Foundry Control Plane?

The **Foundry Control Plane** is a unified management interface that provides:

| Capability | Description |
|------------|-------------|
| **Inventory** | Track all agents, models, and tools across projects |
| **Observability** | Monitor health, performance, and cost metrics |
| **Compliance** | Enforce guardrail policies and security controls |
| **Governance** | Manage quotas, permissions, and access |

## Agent Registry: Discovering & Registering Agents

The Control Plane automatically discovers agents from supported platforms:

### Auto-Discovered Agent Types
- **Foundry Agents** - Prompt-based agents, workflows, hosted agents
- **Azure SRE Agent** - AI-powered Azure resource monitoring
- **Azure Logic Apps** - Agent loop workflows

### Registering Custom Agents

For agents running on external infrastructure (Azure compute, other clouds), you can **register them manually** to gain visibility and control:

1. **Navigate**: `Operate â†’ Overview â†’ Register agent`
2. **Configure**:
   - Agent URL (endpoint where agent receives requests)
   - Protocol (HTTP or A2A)
   - OpenTelemetry Agent ID (for trace correlation)
3. **Select Project**: Choose project with AI Gateway enabled
4. **Save**: Agent appears in Assets inventory

> ðŸ’¡ **Key Benefit**: Once registered, Foundry acts as a proxy via AI Gateway, enabling governance features like blocking/unblocking and telemetry collection.

## Agent Inventory View

The **Assets â†’ Agents** tab provides a unified view of all agents across your subscription:

| Column | Description |
|--------|-------------|
| Name | Agent name or resource name |
| Source | Platform (Foundry, Azure SRE, Logic Apps, Custom) |
| Status | Running, Stopped, Blocked, Unknown |
| Version | Agent version (for Foundry agents) |
| Error Rate | Failed runs percentage (last month) |
| Estimated Cost | Token-based cost estimate |
| Token Usage | Tokens consumed (last month) |
| Runs | Execution count (last month) |

### Lifecycle Operations

| Platform | Supported Actions |
|----------|------------------|
| Foundry (published) | Start / Stop |
| Foundry (hosted, unpublished) | Start / Stop |
| Azure SRE | Start / Stop |
| Logic Apps | Start / Stop (stops entire Logic App) |
| Custom | Block / Unblock |

## Monitoring & Observability

Control Plane leverages **Azure Application Insights** for comprehensive monitoring.

### What's Tracked
- Fleet health metrics (active agents, completion rates, errors)
- Cost and performance (token usage, budget consumption)
- Anomaly detection (cost spikes, performance degradation)
- Drill-down traces (from fleet-level to individual runs)

### Configuration Requirements
1. Azure Application Insights connected to your Foundry project
2. Agents instrumented with OpenTelemetry (GenAI semantic conventions)
3. Appropriate RBAC roles (Log Analytics Reader, Cost Management Reader)


## Compliance & Guardrail Policies

Enforce enterprise-wide guardrails for safety, compliance, and quality.

### Creating Guardrail Policies

1. **Navigate**: `Operate â†’ Compliance â†’ Create policy`
2. **Select Controls**:
   - Content safety filters
   - Prompt shields
   - Groundedness checks
   - Abuse monitoring
3. **Set Scope**: Subscription or resource group
4. **Add Exceptions**: Exclude specific deployments if needed
5. **Deploy**: Policy enforces minimum guardrail requirements

### Compliance Monitoring

| Tab | View |
|-----|------|
| Policies | All policies with compliance status |
| Assets | Model deployments with violation indicators |
| Guardrails | Compare configurations across deployments |
| Security | Defender/Purview integration status |

## AI Gateway Integration

AI Gateway (powered by Azure API Management) enables advanced governance:

### Capabilities
- **Token Rate Limiting**: Enforce TPM (tokens-per-minute) limits
- **Quota Management**: Set total token quotas per period
- **Custom Agent Proxy**: Route traffic through managed gateway
- **Tool Governance**: Control MCP and A2A agent tools

### Setup Steps
1. `Operate â†’ Admin console â†’ AI Gateway tab`
2. Select Foundry resource
3. Create new or use existing APIM instance
4. Enable projects for AI Gateway

> ðŸ’¡ AI Gateway is **free for the first 100,000 API requests**.

## Security Integration

Control Plane integrates with Microsoft security ecosystem:

| Service | Integration |
|---------|------------|
| **Microsoft Defender** | Security posture recommendations, threat alerts |
| **Microsoft Purview** | Data security, compliance, audit logging |
| **Microsoft Entra** | Agent identity management |

### Enabling Security Features

1. **Defender for Cloud**: `Operate â†’ Compliance â†’ Security â†’ Enable`
2. **Purview**: `Operate â†’ Compliance â†’ Security â†’ Enable Microsoft Purview toggle`

Once enabled, security recommendations and alerts surface directly in the Control Plane dashboard.
